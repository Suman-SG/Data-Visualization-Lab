```python
import matplotlib.pyplot as plt

df.head(10).plot(kind='bar', y='median_house_value')
plt.title("Bar Plot")
plt.show()
```

```python
plt.scatter(df['housing_median_age'], df['latitude'])
plt.title("Scatter Plot")
plt.xlabel("Income")
plt.ylabel("House Value")
plt.show()
```

```python
data = df['housing_median_age'].head(5)

plt.pie(data, labels=data.index, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
plt.plot(df['median_income'].head(100))
plt.title("Line Plot")
plt.show()
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(df[['housing_median_age','median_house_value','population']], figsize=(8,8))
plt.show()
```
