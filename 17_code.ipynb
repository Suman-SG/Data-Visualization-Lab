```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 1. Load the dataset
# Adjust path if necessary
df = pd.read_csv('/content/sample_data/california_housing_test.csv')

# 2. Scale all the values in the table
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Save the scaled data to a new CSV file
df_scaled.to_csv('california_housing_scaled.csv', index=False)

# 3. Create 1 box graph with 2 "good" features
# We select 'median_income' and 'median_house_value'
features_to_plot = ['median_income', 'median_house_value']

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_scaled[features_to_plot])

plt.title('Boxplot of Scaled Median Income and Median House Value')
plt.ylabel('Value (Standard Deviations)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('housing_boxplot.png')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Generate Synthetic Data
np.random.seed(42)

# Main Trend: Positive Correlation (Experience vs Salary)
experience = np.linspace(1, 20, 100)
salary = 3000 + (experience * 500) + np.random.normal(0, 500, 100)

# 2. Add a Cluster (A group of interns/entry-level with low variance)
cluster_exp = np.random.normal(2, 0.5, 20)
cluster_sal = np.random.normal(2500, 200, 20)

# 3. Add Outliers (High earners with low experience)
outlier_exp = np.array([1, 2, 3])
outlier_sal = np.array([15000, 14000, 16000])

# Combine into a DataFrame
df = pd.DataFrame({
    'Experience': np.concatenate([experience, cluster_exp, outlier_exp]),
    'Salary': np.concatenate([salary, cluster_sal, outlier_sal])
})

# 4. Create the Plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Experience', y='Salary', s=80, alpha=0.7, color='teal')

# Annotations for clarity
plt.annotate('Outliers', xy=(2, 15000), xytext=(5, 16000),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('Entry-Level Cluster', xy=(2, 2500), xytext=(6, 1000),
             arrowprops=dict(facecolor='red', shrink=0.05))

plt.title('Scatter Plot: Experience vs Salary (Correlation, Clusters, & Outliers)', fontsize=15)
plt.xlabel('Years of Experience', fontsize=12)
plt.ylabel('Salary ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
